```python
CompanyPairSentence_txt=[]
with open("../OriginalDataSet/CompanyPairSentence.txt",'r',encoding="utf-8") as f:
    for row in f:
        row=row.replace('cut','\t')
        row_split = row.split('\t') 
        CompanyPairSentence_txt.append(row_split)
```

```python
CompanyPairSentence_txt
```

```python
CompanyPairSentenceContent=[]
CompanyPairSentenceId=[]
CompanysInSentence=[]
for row in CompanyPairSentence_txt:  
    CompanyPairPerSentence=[] 
    sentence=row[-1] 
    CompanyPairSentenceContent.append(sentence)
    sentenceid=row[0]
    CompanyPairSentenceId.append(sentenceid)
    for company in row[1:-1]:
        CompanyPairPerSentence.append(company)
    CompanysInSentence.append(CompanyPairPerSentence)
```

```python
CompanysInSentence
```

```python
for i in range(len(CompanyPairSentenceContent)):
    for j in range(len(CompanyPairSentenceContent[i])):
        try:
            CompanyPairSentenceContent[i][j].encode('gbk')
        except:
            CompanyPairSentenceContent[i]=CompanyPairSentenceContent[i].replace(CompanyPairSentenceContent[i][j],' ')
    CompanyPairSentenceContent[i]=CompanyPairSentenceContent[i].replace(' ','')
```

```python
file = open('CompanyPairSentenceContent.txt', 'w')
for i in range(len(CompanyPairSentenceContent)):
    file.write(CompanyPairSentenceContent[i])
file.close()
```
