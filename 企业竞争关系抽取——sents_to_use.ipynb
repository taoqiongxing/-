```python
#coding=utf-8

def cut_sentence_new(words):
    # words = (words).decode('utf8')
    start = 0
    i = 0
    sents = []

    punt_list = '。！？～ \t'
    for word in words:
        if word in punt_list and token not in punt_list: #检查标点符号下一个字符是否还是标点
            sents.append(words[start:i+1])
            start = i+1
            i += 1
        else:
            i += 1
            token = list(words[start:i+2]).pop() # 取下一个字符
    if start < len(words):
        sents.append(words[start:])
    return sents
datas_sents=cut_sentence_new(datas)
```

```python
def cut_names(name):
    start = 0
    i = 0
    names = []

    punt_list = ' '
    for word in name:
        if word in punt_list and token not in punt_list: #检查标点符号下一个字符是否还是标点
            names.append(name[start:i+1])
            start = i+1
            i += 1
        else:
            i += 1
            token = list(name[start:i+2]).pop() # 取下一个字符
    if start < len(name):
        names.append(name[start:])
    return names

companys_names=[]
for i in range(0,num_companys):
    cn=cut_names(companys.companys[i])
    cn.sort(key = lambda i:len(i),reverse=True)
    companys_names.append(cn)
```
