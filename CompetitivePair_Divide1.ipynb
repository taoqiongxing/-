```python
import pandas as pd
file = open('../../OriginalDataSet/CompetitivePair.txt','rb')
CompetitivePair=pd.read_csv(file)
file.close()
```

```python
Companys_train=[]
file = open('../../CompanyName/CompanysNames_train.txt')
for line in file.readlines():
    line=line.strip()
    names=line.split(' ')
    Companys_train.append(names[0])
file.close()
```

```python
Companys_val=[]
file = open('../../CompanyName/CompanysNames_val.txt')
for line in file.readlines():
    line=line.strip()
    names=line.split(' ')
    Companys_val.append(names[0])
file.close()
```

```python
Companys_val
```

```python
Companys_test=[]
file = open('../../CompanyName/CompanysNames_test.txt')
for line in file.readlines():
    line=line.strip()
    names=line.split(' ')
    Companys_test.append(names[0])
file.close()
```

```python
Companys_test
```

```python
CompetitivePair_all=[]
CompetitivePair_train=[]
CompetitivePair_val=[]
CompetitivePair_test=[]

    
for i in range(len(CompetitivePair)):
    pair=[]
    pair.append(CompetitivePair.company1[i])        
    pair.append(CompetitivePair.company2[i])
    CompetitivePair_all.append(pair)
    if CompetitivePair.company1[i] not in Companys_val and CompetitivePair.company2[i] not in Companys_val:
        if CompetitivePair.company2[i] not in Companys_test and CompetitivePair.company2[i] not in Companys_test:
            CompetitivePair_train.append(pair)
        else:
            CompetitivePair_test.append(pair)
    elif CompetitivePair.company1[i] not in Companys_test and CompetitivePair.company2[i] not in Companys_test:
        CompetitivePair_val.append(pair)
```
