```python
Companys_val=[]
file = open('../CompanyName/CompanysNames_val.txt')
for line in file.readlines():
    line=line.strip()
    names=line.split(' ')
    Companys_val.append(names[0])
file.close()
```

```python
Companys_test=[]
file = open('../CompanyName/CompanysNames_test.txt')
for line in file.readlines():
    line=line.strip()
    names=line.split(' ')
    Companys_test.append(names[0])
file.close()
```

```python
import pandas as pd
```

```python
CompetitiveSentenceContent_train=[]
CompetitiveSentenceContent_val=[]
CompetitiveSentenceContent_test=[]

CompetitiveSentenceId_train=[]
CompetitiveSentenceId_val=[]
CompetitiveSentenceId_test=[]

CompanysInCompetitiveSentence_train=[]
CompanysInCompetitiveSentence_val=[]
CompanysInCompetitiveSentence_test=[]

for i in range(len(CompetitiveSentenceContent)):
        
    companysinsentence=pd.DataFrame(CompanysInCompetitiveSentence[i])
        
    companysinsentence_val=companysinsentence.isin(Companys_val)
    companysinsentence_test=companysinsentence.isin(Companys_test)

    companysinsentence_val.columns=['isin']
    companysinsentence_test.columns=['isin']
    
    valisin=sum(companysinsentence_val['isin']) 
    testisin=sum(companysinsentence_test['isin'])

    if valisin==0 and testisin==0:
        CompetitiveSentenceContent_train.append(CompetitiveSentenceContent[i])
        CompetitiveSentenceId_train.append(CompetitiveSentenceId[i])
        CompanysInCompetitiveSentence_train.append(CompanysInCompetitiveSentence[i])
        
    if valisin>0 and testisin==0:
        CompetitiveSentenceContent_val.append(CompetitiveSentenceContent[i])
        CompetitiveSentenceId_val.append(CompetitiveSentenceId[i])
        CompanysInCompetitiveSentence_val.append(CompanysInCompetitiveSentence[i])
        
    if valisin==0 and testisin>0:
        CompetitiveSentenceContent_test.append(CompetitiveSentenceContent[i])
        CompetitiveSentenceId_test.append(CompetitiveSentenceId[i])
        CompanysInCompetitiveSentence_test.append(CompanysInCompetitiveSentence[i])   
```
