```python
import jieba
from jieba import posseg

#添加需要自定以的分词  
def addwordslist(filepath):  
    addwords = [line.strip() for line in open(filepath,encoding="utf8").readlines()]  
    return addwords 

addwords = addwordslist('C:/Users/Ax/Desktop/111111/companyname2.txt')
for word in addwords:
    jieba.add_word(word, tag="companyname", freq =1000000000000)
    
import re
# 创建停用词list  
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath,encoding="utf8").readlines()]  
    return stopwords 

def dt_remove(data):
    pattern = u'[a-zA-Z0-9]+'
    wd_length = len(re.findall(pattern,data))
    return wd_length

#定义个函数式用于分词
def jiebaclearText(text):
    sentence_seged = jieba.cut(text)
    stopwords = stopwordslist('C:/Users/Ax/Desktop/111111/stopwords.txt')  # 这里加载停用词的路径  
    outstr = []
    
    for word in sentence_seged: 
        if word not in stopwords:                          
            if word != '\t':  
                if word != '\xa0':
                    if dt_remove(word) ==0:
                        outstr.append(word) 
    return outstr  
    
compete_company_pair_sentence_seg=[]    
for i in range(len(compete_company_pair_sentence)):
    compete_company_pair_sentence_seg.append(jiebaclearText(compete_company_pair_sentence.sents[i]))
```

```python
# coding: utf-8
# Copyright (C) 2016 UKP lab
#
# Author: Daniil Sorokin (ukp.tu-darmstadt.de/ukp-home/)
#
# IO for graphs of relations

import json

def load_relation_graphs_from_files(json_files, val_portion=0.0, load_vertices=True):
    """
    Load semantic graphs from multiple json files and if specified reserve a portion of the data for validation.

    :param json_files: list of input json files
    :param val_portion: a portion of the data to reserve for validation
    :return: a tuple of the data and validation data
    """
    print('22222')
    print(json_files)
    data = []
    for json_file in json_files:
        print(json_file)
        with open(json_file) as f:
            if load_vertices:
                data = data + json.load(f)
            else:
                data = data + json.load(f, object_hook=dict_to_graph_with_no_vertices)
    print("Loaded data size:", len(data))

    val_data = []
    if val_portion > 0.0:
        val_size = int(len(data)*val_portion)
        rest_size = len(data) - val_size
        val_data = data[rest_size:]
        data = data[:rest_size]
        print("Training and dev set sizes:", (len(data), len(val_data)))
    return data, val_data

##加载语义图，从json文件以及如果指定要保留一部分数据用于验证。
##输入：输入json文件、用于验证的部分数据
##返回：数据和验证数据的元组
def load_relation_graphs_from_file(json_file, val_portion=0.0, load_vertices=True):
    """
    Load semantic graphs from a json file and if specified reserve a portion of the data for validation.

    :param json_file: the input json file
    :param val_portion: a portion of the data to reserve for validation
    :return: a tuple of the data and validation data
    """
    print('11111')
    return load_relation_graphs_from_files([json_file], val_portion, load_vertices)


def dict_to_graph_with_no_vertices(d):
    if 'vertexSet' in d:
        del d['vertexSet']
    return d
```
